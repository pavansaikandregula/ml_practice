**concept of referencing and why copy function is necessary**

```python
#create an np array
old_array = np.array([1,2,3,4])
#create a new array and it must be assigned to old_array
new_array = old_array
#now change value of new array
new_array[2:] = 100
#check values of new and old array
new_array
#in cpp, you can called at aliasing concept(function reference)
```

```python
old_array
```

```python
#normal one:
x = 20
y =x
y = 30
print(x,y)
```

**we need concept of copy**

```python
new_array = np.copy(old_array)
```

```python
new_array[2:] = 300
print(new_array)
print(old_array)
```
